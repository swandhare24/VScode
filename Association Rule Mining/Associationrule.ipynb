```python
records = []
for i in lines:
    cart = i.replace('\n', '').split(',')
    records.append(cart)
records
```

```python
len(records)
```

### Apriori Algorithm to get rules

```python
sup = (7*5)/len(records)
sup
```

```python
conf = 0.2
lift = 3
```

```python
from apyori import apriori
```

```python
association_rules = apriori(records, min_support=sup, min_confidence=conf, min_lift=lift)
association_rules
```

```python
association_results = list(association_rules)
association_results
```

```python
items = []
sups = []
confs = []
lifts = []
for i in association_results:
    items.append(list(i[0]))
    sups.append(i[1])
    confs.append(i[2][0][2])
    lifts.append(i[2][0][3])
```

```python
association_dct = {'Items':items,
                   'support':sups,
                   'confidence':confs,
                   'lift':lifts}
```

```python
import pandas as pd
df = pd.DataFrame(association_dct)
df
```

```python
df = df.sort_values(by='lift', ascending=False)
df
```

```python
df.to_csv('AssociationResults.csv', index=False)
```
